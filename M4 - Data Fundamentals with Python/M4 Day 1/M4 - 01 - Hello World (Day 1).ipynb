Python will stop running the entire program when it encounters an error (unless the programmer anticipates and handles the error, a more advanced topic we'll cover later in this course).

<h3 id="convert">Converting One Type of Object to Another</h3>


<p>You can change the type of an object in Python; this is called typecasting. For example, you can convert an <i>integer</i> to a <i>float</i> (e.g., 2 to 2.0).</p>
<p>Let's try it out:</p>


```python
# Verify that it is a whole number

type(2)
```

<h4>Casting Integers to Floats</h4>
<p>Let's convert the integer 2 into a float:</p>

```python
float(2)
```

```python
# Convert the integer 2 into a float and check it's type

type(float(2))
```

<p>When we convert an integer to a float, we don't actually change the value (i.e., the meaning) of the number. However, if we convert a float to an integer, we could potentially lose some information. For example, if we convert the float 1.1 into an integer, we will get 1 and lose the decimal information (i.e., 0.1):</p>

```python
# Converting 1.1 to an integer will result in the loss of information:

int(1.1)
```

<h3 id="exp">Expressions</h3>


<p>Expressions in Python can include operations between compatible types (integers and floats). For example, basic arithmetic operations like adding multiple numbers:</p>


```python
# Addition operation expression

43 + 60 + 16 + 41
```

```python
# Subtraction operation expression

50 - 60
```

```python
# Multiplication operation expression

5 * 5
```

```python
# Division operation expression

25 / 5
```

```python
# Division operation expression

25 / 6
```

```python
# Integer division operation expression

25 // 5
```

<h3 id="var">Variables</h3>


<p>Like with most programming languages, we can store values in <i>variables</i> so that we can use them later. For example:</p>


```python
# Store the value in a variable

x = 43 + 60 + 16 + 41
```

<p>To see the value of <code>x</code> in a Google Colab, we can simply place it on the last line of the cell:</p>


```python
# Print the value of the variable

x
```

<p>We can also perform operations on <code>x</code> and store the result in a new variable:</p>


```python
# Use another variable to store the result of the operation between the variable and the value

y = x / 60
y
```

<p>If we store a value in an existing variable, the new value will replace the previous one:</p>


```python
# Overwrite a variable with a new value

x = x / 60
x
```

<p>It's good practice to use meaningful variable names so that you and others can read the code and understand it more easily:</p>


```python
# Give a meaningful name to the variable

total_min = 43 + 42 + 57 # The total duration of various albums in minutes
total_min
```

```python
# Give a meaningful name to the variable

total_hours = total_min / 60 # The total duration of various albums in minutes
total_hours
```

<p>In the cells above, we added the duration of three albums in minutes and stored it in <code>total_min</code>. Then we divided it by 60 to calculate the total duration, <code>total_hours</code>, in hours. You can also do it all in one expression, as long as you use parentheses to sum the duration of the albums before dividing, as shown below.</p>


```python
# Complicated expression

total_hours = (43 + 42 + 57) / 60  # Total hours in a single expression
total_hours
```

<p>If you prefer to have the total hours as an integer, you can replace floating-point division with integer division (e.g., <code>//</code>).</p>
