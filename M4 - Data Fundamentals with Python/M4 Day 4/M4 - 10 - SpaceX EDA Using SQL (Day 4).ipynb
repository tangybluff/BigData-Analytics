## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
#Rellena la variable query con tu consulta
# SELECT column_name
# FROM table_name
# WHERE column_name LIKE 'variable%';

# 'query' contains an SQL query that will be executed against a database
#  triple double quotes are often used for multi-line strings
query = """  SELECT *
             FROM SPACEXTBL
             WHERE LaunchSite LIKE 'CCS%'
             LIMIT 5;  """
df  = pd.read_sql(query, con = engine )
df

# If I want to select a range of rows in two columns:
# SELECT column1, column2 or *
# FROM table_name
# LIMIT offset, count;
# 'offset' is the starting row number (0-based index) from where you want to begin fetching rows
# 'count' is the number of rows you want to retrieve
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
# Customers
# PayloadMass

query = """  SELECT FlightNumber, SUM(PayloadMass) AS TotalPayloadMass
              FROM SPACEXTBL
              WHERE Customers = 'NASA (CRS)'
              GROUP BY FlightNumber;  """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """ SELECT ROUND(AVG(PayloadMass))
            FROM SPACEXTBL
            WHERE serial LIKE 'B1%' ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
# Most efficient form:
query = """ SELECT FlightNumber, BoosterVersion, LandingOutcome, MIN(Date)
FROM SPACEXTBL
WHERE LandingOutcome = 'True RTLS'; """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """ SELECT BoosterVersion, LandingOutcome, PayloadMass
FROM SPACEXTBL
WHERE LandingOutcome = 'True ASDS' AND PayloadMass BETWEEN 4000 AND 6000 ; """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = """
SELECT Mission_Outcome AS 'Mission Value',
  CASE WHEN Mission_Outcome = 0 THEN 'Mission Failed'
       WHEN Mission_Outcome = 1 THEN 'Mission Accomplished'
  END AS 'Mission Status',
  COUNT(*) AS 'Total'
FROM SPACEXTBL
GROUP BY Mission_Outcome ; """
df = pd.read_sql(query, con = engine)
df

# From my query, I've given the column 'Mission_Outcome' an alias of 'Mission Outcome'
# Whatever I put in my CASE are my options and END AS is the name of the column the two CASE WHEN (x2)
# COUNT(*) is used to count all the rows or records in the table
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
# What we need to do is first find the maximum mass that any rocket has ever taken
# And then, we see which rockets, by their serial number, carried that maximum payload weight..
query = """
SELECT FlightNumber, Serial, PayloadMass AS 'Max. Mass'
FROM SPACEXTBL
WHERE PayloadMass = (
  SELECT MAX(PayloadMass)
  FROM SPACEXTBL); """
df  = pd.read_sql(query, con = engine )
df
```
