### BIG DATA: Media aritm√©tica y varianza

```python
import numpy as np
import pandas as pd # Manipulation for dataframes
import matplotlib.pyplot as plt # To make tables
```

```python
# Crear los datos de columna x1; this data is invented 
datos = {'x1':[10.2, 10.6, 10.4, 10.8, 10, 9.8, 9.4, 9.6, 9.2, 10]}
# Crear el dataframe
df = pd.DataFrame(datos)
# Crear los datos de columna x2
df["x2"] = [11, 13, 12, 14, 10, 9, 7, 8, 6, 10]
# echar un vistazo
df
```

```python
# 'punto de partida' ; this is for basic plotting and you can copy and paste it
fig = plt.figure(figsize=(14, 8))
plt.plot(df.index, df["x1"], color="blue", marker="o")
plt.plot(df.index, df["x2"], color="orange", marker="o")
plt.axhline(y = df["x2"].mean(), color = 'r', linestyle = '-')
plt.xticks([])
plt.yticks(fontsize=20)
plt.legend(["x1", "x2","media aritm√©tica"], loc ="upper right", fontsize=22);
```

```python
# This is a function of pandas at the end; we do this for the column x1
print("Media aritm√©tica de x1:   ", df["x1"].mean())
print("varianza de x1:           ", df["x1"].var())
print("desviaci√≥n est√°ndar de x1:", df["x1"].std())
```

```python
# This is done for column x2
print("Media aritm√©tica de x2:   ", df["x2"].mean())
print("varianza de x2:           ", df["x2"].var())
print("desviaci√≥n est√°ndar de x2:", df["x2"].std())
```
